```python
import matplotlib.pyplot as plt

def temp_analysis(df):

    daily_temp = df.groupby("region")["temperature_celsius"].sum()

    avg_temp = df.groupby("region")["temperature_celsius"].mean()

    daily_temp_infahrenheit = df.groupby("region")["temperature_fahrenheit"].sum()

    return daily_temp, avg_temp, daily_temp_infahrenheit


daily_temp, avg_temp, daily_temp_infahrenheit = temp_analysis(df)

# Line plot for daily temperature
plt.figure()

plt.bar(daily_temp.index, daily_temp.values)

plt.title("Total Temperature by Timezone")
plt.xlabel("Region")
plt.ylabel("Temperature Celsius")
plt.xticks(rotation=45)

plt.show()


# Histogram for average temperature
plt.figure()
plt.hist(avg_temp.values, bins=30)

plt.title("Average Temperature in City")
plt.xlabel("Average Temperature")
plt.ylabel("Frequency")
plt.show()


# Fahrenheit trend
plt.figure()

plt.bar(daily_temp_infahrenheit.index, daily_temp_infahrenheit.values)

plt.title("Temperature in Fahrenheit by Region")
plt.xlabel("Region")
plt.ylabel("Temperature Fahrenheit")
plt.xticks(rotation=45)

plt.show()
```
